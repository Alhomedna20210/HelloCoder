```python
import requests
from bs4 import BeautifulSoup
import os
import urllib.request

# Function to create directory if it doesn't exist
def create_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

# Function to scrape images from a website
def scrape_images(url, directory):
    # Send a GET request to the URL
    response = requests.get(url)
    # Parse HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    # Find all image tags
    image_tags = soup.find_all('img')
    # Iterate through image tags
    for img in image_tags:
        # Get the source URL of the image
        image_url = img.get('src')
        # Download the image
        try:
            urllib.request.urlretrieve(image_url, f"{directory}/{image_url.split('/')[-1]}")
            print(f"Downloaded: {image_url}")
        except Exception as e:
            print(f"Failed to download: {image_url}")
            print(e)

# URL of the website to scrape images from
url = "https://unsplash.com/s/photos/human-faces"
# Directory to save the downloaded images
directory = "images"

# Create the directory if it doesn't exist
create_directory(directory)
# Scrape images from the website
scrape_images(url, directory)

# Scrape images without specifying the number
scrape_images(url, directory)

from google.colab import drive
drive.mount('/content/drive')

import os
os.chdir("/content/drive/My Drive/images")
```
